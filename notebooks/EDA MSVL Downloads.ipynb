```python
import pandas as pd
from datetime import datetime
```

```python
df_trend = pd.read_csv("data/MSMVL_trend_90d.csv",header=1)
```

```python
df_trend.head(3)
```

```python
df_trend['Day'] = df_trend['Day'].map(lambda x: datetime.strptime(x,'%Y-%m-%d').date())
```

```python
df_trend.head(10)
```

```python
df_stock = pd.read_csv("data/MASe_Daily_stock.csv",header=0)
```

```python
df_stock.head(3)
```

```python
df_stock['Date'] = df_stock['Date'].map(lambda x: datetime.strptime(x,'%b %d, %Y').date())
```

```python
df_stock.head(3)
```

```python
df_full=pd.merge(df_trend,df_stock,right_on='Date', left_on='Day', how='left')
```

```python
df_full.head(10)
```

```python
def plot(df,x1,y1,y2):
    import plotly.offline as py
    py.init_notebook_mode(connected=True)
    import plotly.graph_objs as go
    from datetime import datetime
    trace1 = go.Scatter(x=df[x1], 
                            y=df[y1],
                            name = y1,
                            line = dict(color = '#97ef9e'),
                            opacity = 0.8)
    trace2 = go.Scatter(x=df[x1],
                           y=df[y2],
                           name = y2,
                           line = dict(color = '#f27d60'),
                           opacity = 0.4)
    data = [trace1,trace2] 
    
    fig = dict(data = data)
    py.iplot(fig)
    
plot(df_full,'Day','Price','masmovil: (Spain)')
```
