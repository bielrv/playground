```python
layout = dict(
            xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1,
                         label='1m',
                         step='month',
                         stepmode='backward'),
                    dict(count=3,
                         label='3m',
                         step='month',
                         stepmode='backward'),
                    dict(count=6,
                         label='5m',
                         step='month',
                         stepmode='backward')
                ])
            ),
            rangeslider=dict(
                visible = True
            ),
            type='date'
        ),
    yaxis=dict(
        title='Trend axis'
    ),
    yaxis2=dict(
        title='Stock axis',
        titlefont=dict(
            color='rgb(148, 103, 189)'
        ),
        tickfont=dict(
            color='rgb(148, 103, 189)'
        ),
        overlaying='y',
        side='right')
    )
```

```python
def plot_StockPrices(layout: dict):

    
    import plotly.offline as py
    py.init_notebook_mode(connected=True)
    
    import plotly.graph_objs as go
    from datetime import datetime
    
    MasMovil_trend = go.Scatter(x=df.index, 
                            y=df.MasMovil,
                            name = 'MasMovil trend',
                            line = dict(color = '#97ef9e'),
                            opacity = 0.8,
                            yaxis='y2')
    
    Vodafone_trend = go.Scatter(x=df.index,
                           y=df.Vodafone,
                           name = 'Vodafone trend',
                           line = dict(color = '#f27d60'),
                           opacity = 0.4,
                            yaxis='y2')
    
    MasMovil_stock = go.Scatter(x=df.index,
                       y=df['4. close'],
                       name = 'MasMovil stock')
    
    MasMovil_volume = go.Scatter(x=df.index,
                   y=df['5. volume'],
                   name = 'MasMovil volume')
    
    
    data = [MasMovil_trend, MasMovil_stock,MasMovil_volume] 
    
    layout.update({'title': 'Stock & Trends'})
    
    fig = dict(data = data ,layout = layout)
    
    py.iplot(fig)
```

```python
plot_StockPrices(layout)
```
